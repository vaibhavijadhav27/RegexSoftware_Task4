```python
plt.title('Comparison between sepal width and length')
sns.scatterplot(iris['SepalLengthCm'], iris['SepalWidthCm']);
```

```python
plt.title('Comparison between sepal width and length on the basis of species')
sns.scatterplot(iris['SepalLengthCm'], iris['SepalWidthCm'], hue = iris['Species'], s= 50);
```

```python
plt.title('Comparison between petal width and length')
sns.scatterplot(iris['PetalLengthCm'], iris['PetalWidthCm']);
```

```python
plt.title('Comparison between petal width and length on the basis of species')
sns.scatterplot(iris['PetalLengthCm'], iris['PetalWidthCm'], hue = iris['Species'], s= 50);
```

```python
sns.pairplot(iris,hue="Species",height=3);
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr(), annot = True);
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))
sns.boxplot(  y="PetalWidthCm", x= "Species", data=iris,  orient='v' , ax=axes[0, 0])
sns.boxplot(  y="PetalLengthCm", x= "Species", data=iris,  orient='v' , ax=axes[0, 1])
sns.boxplot(  y="SepalWidthCm", x= "Species", data=iris,  orient='v' , ax=axes[1, 0])
sns.boxplot(  y="SepalLengthCm", x= "Species", data=iris,  orient='v' , ax=axes[1, 1])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))
sns.violinplot(y="PetalWidthCm", x= "Species", data=iris,  orient='v' , ax=axes[0, 0])
sns.violinplot(y="PetalLengthCm", x= "Species", data=iris,  orient='v' , ax=axes[0, 1])
sns.violinplot(y="SepalWidthCm", x= "Species", data=iris,  orient='v' , ax=axes[1, 0])
sns.violinplot(y="SepalLengthCm", x= "Species", data=iris,  orient='v' , ax=axes[1, 1])
plt.show()
```
